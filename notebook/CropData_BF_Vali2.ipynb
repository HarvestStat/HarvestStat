```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import proplot
```

```python
df = pd.read_excel('/Users/dlee/Downloads/BF_Ag_Apr272024.xlsx', sheet_name='Data')
df.loc[df['Admin 2'].isna(), 'name'] = df['Admin 1']
df.loc[df['Admin 2'].notna(), 'name'] = df['Admin 2']
```

```python
# Kadiogo (Millet)
name, product_name, fnid_list = 'Kadiogo', 'Mil', ['BF1997A122', 'BF2001A20301','BF2020A21300']
sub = df[df['FNID'].isin(fnid_list)]
sub = sub[(sub['DNL Source Crop'] == product_name)]
sub = sub.pivot_table(index='Year', columns=['FNID','name'], values='Planted: ha', aggfunc='sum').reindex(np.arange(1984, 2023), fill_value=np.nan)
sub.sort_index(inplace=True)
# Plot
fig, ax = plt.subplots(figsize=(7,3))
fig.set_facecolor('white')
sub.plot(ax=ax, marker='o', markersize=5, title='%s - %s' % (name, product_name))
ax.legend().set_title('')
ax.set_xlabel('')
ax.set_ylabel('Area Planted (ha)')
plt.tight_layout()
plt.show()
```
